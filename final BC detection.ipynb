```python
#extract only the mean calculations of the features and generating a heatmap to better see the correlation

features_mean = df.iloc[:, :11]
print(features_mean.corr())
plt.figure(figsize = (13,13))
sns.heatmap(features_mean.corr(),annot = True, cmap = 'coolwarm')
```

```python
#encodeing the diagnosis column

from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
features_mean['diagnosis'] = encoder.fit_transform(features_mean['diagnosis'])
features_mean['diagnosis']
```

```python
#extracting the features in X and the labels in Y

X = features_mean.drop(['diagnosis'], axis=1)
Y = features_mean['diagnosis']
```

```python
#splitting the data into training and testing datasets

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
# define a function that trains different models on the data

def models(X_train, Y_train):
    
    # logistic regression model
    from sklearn.linear_model import LogisticRegression
    lrm = LogisticRegression(random_state=0)
    lrm.fit(X_train, Y_train)
    
    # decision tree
    from sklearn.tree import DecisionTreeClassifier
    dtc = DecisionTreeClassifier(criterion='entropy', random_state=0)
    dtc.fit(X_train, Y_train)
    
    #Random forest classifier
    from sklearn.ensemble import RandomForestClassifier
    rfc = RandomForestClassifier(n_estimators=22, criterion='entropy', max_depth=10, random_state=42)
    rfc.fit(X_train, Y_train)
    
    #support vector machine
    from sklearn.svm import SVC
    svc = SVC(kernel='linear', C = 1)
    svc.fit(X_train, Y_train)
    
    #xgboost
    from xgboost import XGBClassifier
    xgb = XGBClassifier(n_estimator = 20)
    xgb.fit(X_train, Y_train)
    
    #naive bayes
    from sklearn.naive_bayes import GaussianNB
    gnb = GaussianNB()
    gnb.fit(X_train, Y_train)
    
    #print the model accuracy on the training data
    print('[0] LogisticRegression training accuracy: ', lrm.score(X_train, Y_train))
    print('[1] DecisionTreeClassifier training accuracy: ', dtc.score(X_train, Y_train))
    print('[2] RandomForestClassifier training accuracy: ', rfc.score(X_train, Y_train))
    print('[3] SVC training accuracy: ', svc.score(X_train, Y_train))
    print('[4] XGBClassifier training accuracy: ', xgb.score(X_train, Y_train))
    print('[5] GaussianNB training accuracy: ', gnb.score(X_train, Y_train))
    return lrm, dtc, rfc, svc, xgb, gnb
```

```python
#traing the models

model = models(X_train, Y_train)
```

```python
#evaluating the model performance

from sklearn.metrics import classification_report, accuracy_score
for i in range(len(model)):
    print(f'model {i}')
    print(classification_report(Y_test, model[i].predict(X_test)))
    print(accuracy_score(Y_test, model[i].predict(X_test)))
```

```python
#plotting the confusion matrix 

from sklearn.metrics import plot_confusion_matrix

for i in range(len(model)):
    plot_confusion_matrix(model[i], X_test, Y_test)
```
